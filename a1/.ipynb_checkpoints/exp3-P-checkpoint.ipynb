```python
import numpy as np
import matplotlib.pyplot as plt

def exp3p(d,T,eta,beta):
    e_loss = 0
    elv = 0.5*np.ones([d,2])
    elv[-2,:] = 0.4
    elv[-1,:] = [0.6,0.3]
    egv = 1-elv
    w_tilde = np.ones([d])

    for t in range(T):
        w = w_tilde/np.sum(w_tilde)
        adv_choice = np.random.choice(d,p=w)
        e_gain_c   = egv[adv_choice,(2*t)//T]
        gain       = beta*np.ones([d])/
        l          = np.random.choice(2,p=[1-e_loss_c, e_loss_c])/w[adv_choice]
        e_loss    += e_loss_c
        w_tilde[adv_choice]    = w_tilde[adv_choice]*np.exp(-eta*l)

    return e_loss - 0.4*T
```
